Q.No-1
a)What is Matplotlib?
    Matplotlib was created by John D. Hunter.
    Matplotlib is a low level graph plotting library in python that serves as a visualization utility.
b)Why is it used?
    Matplotlib is a python library which is used for visualization and analysis purpose and it is also used for creating attractive graphs and charts.
c)Name five plots that can be plotted using the Pyplot module of Matplotlib.
    *Scatter plot: Dot-based plotting of multiple variables along x and y axes represents a scatter plot.
    *Pie Chart: A Pie plot is a circular graph where the data get represented within that components/segments or slices of pie.
    *Line plot: Line plots are drawn by joining straight lines connecting data points where the x-axis and y-axis values interesect.
    *Box plot: A Box Plot is the visual representation of the statistical five number summary of a given data set.
    *Subplot() function can be called to plot multiple plots in the same figure.

```python
#Q.No-2
#Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
x=np.array([10,20,30,40,50,60,70])
y=np.array([100,200,330,400,550,600,700])
plt.scatter(x,y)
```

```python
#Q-N0-3
#Draw four line plots using the subplot() function.
x=np.array([1,2,3,4,5])
y1=np.array([30,45,64,32,56])
y2=np.array([11,43,89,43,22])
y3=np.array([23,12,11,34,7])
y4=np.array([22,81,29,54,37])
plt.subplot(1,4,1)
plt.plot(x,y1,'blue')
plt.subplot(1,4,2)
plt.plot(x,y2,'pink')
plt.subplot(1,4,3)
plt.plot(x,y3,'black')
plt.subplot(1,4,4)
plt.plot(x,y4,'red')
```

```python
#Q-4
#Using the following data plot a bar plot and a horizontal bar plot.
x=np.array(['A','B','C','D','E'])
y=np.array([10,20,30,40,50])
plt.bar(x,y,color='y')
```

```python
x=np.array(['A','B','C','D','E'])
y=np.array([10,20,30,40,50])
plt.barh(x,y,color='y')
```

```python
#Q.No-5
#Using the following data plot a box plot.
x=[25,30,35,40,45,50,55,60,65,70]
plt.boxplot(x)
```

Seaborn

a)What is Seaborn?
    Seaborn is a well-known Python library for data visualization that offers a user-friendly interface for producing visually appealing and informative statistical graphics. It is designed to work with Pandas dataframes, making it easy to visualize and explore data quickly and effectively.

Here are some of the most commonly used plot types in Seaborn:
    *Scatter Plot: A scatter plot is used to visualize the relationship between two variables. Seaborn's scatterplot() function provides a simple way to create scatter plots.
    *Line Plot: A line plot is used to visualize the trend of a variable over time. Seaborn's lineplot() function provides a simple way to create line plots.

    *Histogram: A histogram is used to visualize the distribution of a variable. Seaborn's histplot() function provides a simple way to create histograms.

    *Box Plot: A box plot is used to visualize the distribution of a variable. Seaborn's boxplot() function provides a simple way to create box plots.

    *Heatmap: A heatmap is used to visualize the correlation between different variables. Seaborn's heatmap() function provides a simple way to create heatmaps.


```python
titanic=pd.read_csv('titanic.csv')
titanic
```

```python
import seaborn as sns
sns.boxplot(x='pclass',y='age',data=titanic)
```

```python
diamonds=pd.read_csv('diamonds.csv')
diamonds
```

```python
sns.histplot(data=diamonds,x="price",hue="cut")
```

```python
iris=pd.read_csv('iris.csv')
iris
```

```python
sns.pairplot(data=iris,hue="species")
```

```python
flights=pd.read_csv('flights.csv')
flights
```

```python
flights_pivot=flights.pivot_table(index="month",columns="year",values="passengers")
sns.heatmap(flights_pivot,annot=True,fmt="d",linewidths=.5)
```
